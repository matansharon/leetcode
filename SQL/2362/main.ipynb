```python
import pandas as pd
import sqlalchemy
import pymysql
```

```python
conn = sqlalchemy.create_engine("mysql+pymysql://root:ma291188@localhost/leetcode")
products=pd.read_sql_table(con=conn,table_name="products")
orders=pd.read_sql_table(con=conn,table_name="orders")
```

```python
orders
```

```python
dict_of_orders={}
for i in range(len(orders)):
    row=orders.iloc[i]
    if row.order_id not in dict_of_orders:
        dict_of_orders[row.order_id]=[]
    dict_of_orders[row.order_id].append(products.loc[row.product_id].price*row.quantity)
maxes={}
m=0
for i in dict_of_orders:
    s=sum(dict_of_orders[i])
    if s>m:
        maxes={i:s}
        m=s
    if s==m:
        maxes[i]=s
out=pd.DataFrame(maxes,index=range(len(maxes)))
out.to_sql(con=conn,name="out_2362",if_exists="replace")
out.to_csv("res.csv")
```
