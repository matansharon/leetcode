```python
class Date:
    def __init__(self,date_as_string):
        temp=date_as_string.split("/")
        self.year=temp[0]
        self.month=temp[1]
        self.day=temp[2]
    def toString(self):
        return self.year+"/"+self.month+"/"+self.day
```

```python
class Salary:
    def __init__(self,i,emp_id,amt,date):
        self.id=i
        self.emp_id=emp_id
        self.amount=int(amt)
        self.date=Date(date)
    def toString(self):
        return str(self.id)+","+str(self.emp_id)+","+str(self.amount)+","+str(self.date)
        
```
